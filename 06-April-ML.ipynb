```python
## Q No. 2 :
"""
The objective of the support vector machine algorithm is to find a hyperplane in an N-dimensional 
space(N — the number of features) that distinctly classifies the data points.

Possible hyperplanes
To separate the two classes of data points, there are many possible hyperplanes that could be chosen.
Our objective is to find a plane that has the maximum margin, i.e the maximum distance between data points of both classes.
Maximizing the margin distance provides some reinforcement so that future data points can be classified with more confidence.

Hyperplanes and Support Vectors
Hyperplanes are decision boundaries that help classify the data points.
Data points falling on either side of the hyperplane can be attributed to different classes.
Also, the dimension of the hyperplane depends upon the number of features.
If the number of input features is 2, then the hyperplane is just a line. 
If the number of input features is 3, then the hyperplane becomes a two-dimensional plane. 
It becomes difficult to imagine when the number of features exceeds 3.
"""
```

```python
## Q No. 4 :
"""
A support vector machine (SVM) is machine learning algorithm that analyzes data for classification and regression analysis.
SVM is a supervised learning method that looks at data and sorts it into one of two categories.
An SVM outputs a map of the sorted data with the margins between the two as far apart as possible.
"""
```

```python
## Q No. 5 :
"""
Hence, the SVM algorithm helps to find the best line or decision boundary; this best boundary or
region is called as a hyperplane. SVM algorithm finds the closest point of the lines from both the classes. 
These points are called support vectors. 
The distance between the vectors and the hyperplane is called as margin
The difference between a hard margin and a soft margin in SVMs lies in the separability of the data.
If our data is linearly separable, we go for a hard margin. 
However, if this is not the case, it won't be feasible to do that.
"""
```

```python
## Q  No. 6 : 
from sklearn.datasets import load_iris
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
df=pd.DataFrame(data=dataset.data,columns =dataset.feature_names)
```

```python
df.head()
```

```python
df["target"]= dataset.target
```

```python
df.head()
```

```python
X= df.iloc[:,:-1] # independent variable
y=df.iloc[:,-1]   # dependent variable
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```
